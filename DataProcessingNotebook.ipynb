```python
import pandas as pd
import os
import numpy as np
import time
timestr = time.strftime("%Y%m%d-%H%M%S")
# location of a folder containing csv files downloaded from ccc website named per station ids
pathstr = "Data/CycleCountCSV/"
folder = os.fsencode(pathstr)
#makes and empty list. no idea if this does anything but its my emotional support line
counts = []
for file in os.listdir(folder):
    filename=os.fsdecode(file)
    stationid = filename.split('.')
    stationid = stationid[0]
    print(stationid)
    # goes through each file in the folder and sets stationid to the filename without an extension 
    counts.append(pd.read_csv(pathstr+filename, sep=";", names=["date",stationid],dtype={'a':np.datetime64,'b':np.int64}))
    # converts each csv to a pandas dataframe, names the columns date and the station id and declares the datatypes to be datetime65 and int64
counts_joined = counts.pop(0)
#starts it off by taking the first in the list
print(counts_joined)
for frames in counts:
    counts_joined = counts_joined.merge(frames,on="date")
    #iterates through the list of dataframes and merges them together, using the date column as a key
display(counts_joined)
counts_joined.to_csv("Results/mergeddata"+timestr+".csv")
#exports to file
```
